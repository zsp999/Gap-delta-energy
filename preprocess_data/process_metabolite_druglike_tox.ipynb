```python
import matplotlib.pyplot as plt
import pandas as pd
from rdkit import Chem
from rdkit.Chem.SaltRemover import SaltRemover
```

```python
remover = SaltRemover()
def smi_standard(smi):
    try:
        mol = Chem.MolFromSmiles(smi)
        molnew = remover(mol, dontRemoveEverything=True)
        smi = Chem.MolToSmiles(molnew) #moltosmiles起到规范化作用
    except:
        smi = 'None'
        print('error')
    return smi

sfilename = "data/hmdb_metabolites.xml"
strcount1 = 0
strcount2 = 0
smiles = 'None'
average_molecular_weight =0
endo_or_exo ='None'
flag = False
smiles_dict = {'smiles':[], 'standard_smiles':[],'source':[], 'MW':[]}
for line in open(sfilename,'r',encoding='utf-8'):
    line = str(line.strip('\n').strip(' '))
    if line == '</metabolite>':
        flag =True
    if line == '</secondary_accessions>':
        strcount1 =1
        strcount1 =0
    if line[0:26] == '<average_molecular_weight>':
        average_molecular_weight = line[26:-27]
    if line[0:8] == '<smiles>':
        smiles = line[8:-9]
    if line == '<term>Source</term>':
        strcount2 =1
        continue
    if strcount2!= 0:
        strcount2+=1
    if strcount2 == 10:
        endo_or_exo = line[6:-7]
        strcount2=0
    if flag==True:
        smiles_dict['smiles'].append(smiles)
        smiles_dict['standard_smiles'].append(smi_standard(smiles))
        smiles_dict['source'].append(endo_or_exo)
        smiles_dict['MW'].append(average_molecular_weight)
        smiles = 'None'
        average_molecular_weight = 0
        endo_or_exo = 'None'
        flag =False
        if len(smiles_dict['smiles'])%10000 ==0:
            print(len(smiles_dict['smiles']))
```

```python
print(len(smiles_dict['smiles']))
df_meta = pd.DataFrame(smiles_dict)
df_meta['MW'] = df_meta['MW'].apply(lambda x: float(x))
df_meta = df_meta.drop_duplicates('standard_smiles', keep='first') #去重
df_metabolites = df_meta[ (df_meta['MW']<=1000) & (df_meta['standard_smiles'] != 'None')]
df_metabolites.head(10)
```

```python
df_metabolites.to_csv('data/metabolites_small_0108.csv')
```

```python
df_metabolites = pd.read_csv('data/metabolites_small_0108.csv')
key = 'source'
def draw_pie(key, df):
        grades_labels = list(df[key].value_counts().index)
        print(grades_labels)
        grades_nums = df[key].value_counts().values
        plt.figure(figsize=(12,6),dpi=300)
        plt.pie(grades_nums, labels = grades_labels, autopct = '%1.1f%%',
        pctdistance = 1.1,labeldistance = 1.2)
        plt.title(key)
        plt.show()
draw_pie(key, df_metabolites)
```

```python
#处理类药分子
druglike_namelist = ['approved', 'experimental', 
                    'illicit', 'investigate', 'nutraceutical', 'withdrawn']
def precess_druglike(druglike_namelist = druglike_namelist):
    smi_all_dict = {'standard_smiles':[], 'source':[]}
    for drug in druglike_namelist:
        mols = Chem.SDMolSupplier('data/' + drug+ '.sdf')
        print(len(mols))
        smis = []
        remover = SaltRemover()
        for m in range(len(mols)):
            try:
                if float(mols[m].GetProp('MOLECULAR_WEIGHT'))<=1000.0:
                    molnew = remover(mols[m], dontRemoveEverything=True)
                    smi = Chem.MolToSmiles(molnew) #moltosmiles起到规范化作用
                    smis.append(smi)
                else:
                    pass
            except:
                print('error:', mols[m])
        smis = set(smis)
        print('all smiles num:',len(smis))
        smi_all_dict['standard_smiles'].extend(list(smis))
        smi_all_dict['source'].extend([drug for i in range(len(smis))])

    df_drug_smi = pd.DataFrame(smi_all_dict)
    return df_drug_smi

```

```python
df_drug_smi = precess_druglike()
df_drug_smi.to_csv('data/druglike_small_0108.csv')
df_drug_smi.head(-10)
```

```python
tox_namelist = ['toxic']
df_tox_smi = precess_druglike(tox_namelist)
```
